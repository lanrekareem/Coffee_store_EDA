```python
#LETS PLOT THE VARIOUS PRODUCT TYPES BY SALES USING COLUMN CHART
plt.bar(df['Product Type'],df['Sales'])
plt.xlabel('Product Type')
plt.ylabel('Sales')
plt.title('sales by product type')
```

```python
# Sum the profit by market
profit_sum = df.groupby('Market')['Profit'].sum().reset_index()
# Create a bar chart
plt.bar(profit_sum['Market'], profit_sum['Profit'])
plt.xlabel('Market')
plt.ylabel('Total Profit')
plt.title('Total Profit by Market')
plt.show()
```

```python
# Sum the profit by market size
profit_sum = df.groupby('Market Size')['Profit'].sum().reset_index()

plt.bar(profit_sum['Market Size'],profit_sum['Profit'])
plt.xlabel('Market Size')
plt.ylabel('Profit')
plt.title('profit by Market Size')
```

```python
# Select numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()
correlation_matrix
```

```python
#LET PLOT THE FINANCIAL METRICS USING HEATMAP
sns.heatmap(correlation_matrix, annot=True)
plt.rcParams['figure.figsize']=(20,10)
plt.title('Correlation Heatmap of Financial Metrics')
plt.show
```

```python
# Parse the Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
```

```python
#LETS CHECK FOR SALES OVER TIME
monthly_sales = df.set_index('Date').resample('ME')['Sales'].sum()
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales over time')
plt.show()
```

```python
#LETS CHECK FOR PROFIT OVER TIME 
monthly_Profit = df.set_index('Date').resample('ME')['Profit'].sum()
plt.plot(monthly_Profit.index, monthly_Profit.values)
plt.xlabel('Date')
plt.ylabel('Profit')
plt.title('Profit over time')
plt.show()
```
